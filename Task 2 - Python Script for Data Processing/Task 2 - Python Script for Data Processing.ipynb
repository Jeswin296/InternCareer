# Data Processing Script

```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
x = iris.data
y = iris.target

print(x)
```

```python
print(y)
```

```python
np.mean(x)
```

```python
from scipy.stats import norm
```

```python
def histogram(name,data):
    mean = np.mean(data)
    sd = np.std(data)
    x_axis = np.linspace(min(data), max(data),100)
    plt.plot(x_axis,norm.pdf(x_axis, mean, sd), color="Purple", label="Normal Distribution")
    plt.xlabel(f'{name}')
    plt.ylabel('density')
    plt.title(f'Normal Distribution - {name}')
    plt.legend()
    plt.show()
    
    plt.hist(data, bins = 20, density=True, alpha=0.6,color='brown', label = "Data Histogram")
    plt.xlabel(f'{name}')
    plt.ylabel('density')
    plt.title(f'Histogram - {name}')
    plt.legend()
    plt.show()
    
my_sepal_length = np.random.normal(5, 0.5, 150)
my_sepal_width = np.random.normal(3, 0.3, 150)
my_petal_length = np.random.normal(4, 0.8, 150)
my_petal_width = np.random.normal(1.5, 0.2, 150)


histogram("My Sepal Length", my_sepal_length)
histogram("My Sepal Width", my_sepal_width)
histogram("My Petal Length", my_petal_length)
histogram("My Petal Width", my_petal_width)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 99)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

np.mean(x_train_scaled)
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train_scaled, y_train)
lr_predictions = lr_model.predict(x_test_scaled)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train_scaled, y_train)
dt_prediction = dt_model.predict(x_test_scaled)
```

```python
svm_model = SVC()
svm_model.fit(x_train_scaled, y_train)
svm_prediction = svm_model.predict(x_test_scaled)
```

```python
def evaluate_model(model_name, prediction):
    accuracy = accuracy_score(y_test, prediction)
    class_report = classification_report(y_test, prediction, target_names = iris.target_names)
    
    print(f"{model_name} Model")
    print("Accuracy: ", accuracy)
    print("Classification Report:\n ", class_report)
    
evaluate_model("Random forests", lr_predictions)
evaluate_model("Naive Bayes classifiers", svm_prediction)
evaluate_model("Cluster analysis", dt_prediction)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
def plot_confusion_matrix(model_name, prediction):
    cm = confusion_matrix(y_test, prediction)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d",cmap="Oranges",xticklabels = iris.target_names, yticklabels = iris.target_names)
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicated Lables")
    plt.ylabel("True Lables")
    plt.show()
    
plot_confusion_matrix("Random forests", lr_predictions)
plot_confusion_matrix("Naive Bayes classifiers", svm_prediction)
plot_confusion_matrix("Cluster analysis", dt_prediction)
```

```python
def plot_accuracy_comparison(models, accuracies):
    plt.figure(figsize=(8,4))
    plt.bar(models, accuracies, color='green')
    plt.title("Model Accuracy Comparison")
    plt.xlabel("Model")
    plt.ylabel("Accuracy")
    plt.ylim(0.8, 1.0)
    plt.show()
    
model_names = ["Random forests", "Naive Bayes classifiers", "Cluster analysis"]
model_accuracies = [accuracy_score(y_test, lr_predictions),
                  accuracy_score(y_test, svm_prediction),
                  accuracy_score(y_test, dt_prediction)]

plot_accuracy_comparison(model_names, model_accuracies)
```
